Take a look at the metadata.

```
metadata = pd.read_csv('metadata.csv')

f"The dimensions are {metadata.shape}"

metadata.head()
```

Subset the metadata to include only columns that we are interested.

```
metadata.columns
metadata = metadata[['patientid', 'finding', 'sex', 'age', 'location', 'folder', 'date', 'filename', 'modality', 'view']]
```

Filter out the rows that are COVID-19 cases and x-ray scans.

```
metadata = metadata[metadata.finding.eq('COVID-19')]
metadata = metadata[metadata.modality.eq('X-ray')]
metadata.iloc[380:400,:]
```

Get the file names in the images folder and the file names in the metadata.

```
imagefiles = os.listdir("./images")
metadatafiles = metadata['filename'].tolist()
f"Number of images in total: {len(imagefiles)}, number of metadata rows (COVID cases): {len(metadatafiles)}"
```

```
def intersection(lst1, lst2): 
    lst3 = [value for value in lst1 if value in lst2] 
    return lst3 


files = intersection(imagefiles, metadatafiles)
len(files)
```

Move the files into the new folder "COVID19_images". Images have already been copied so there's no need to run this chunk again.
